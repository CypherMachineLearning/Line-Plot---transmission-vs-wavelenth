## __Subsetting the dataframe__

````
# lamda(wavelength)
lam_1 = bk1['lambda(m)']
lam_2 = bk1['lambda(m)']
lam_3 = bk1['lambda(m)']
lam_4 = bk1['lambda(m)']

# transmission
y_1 = bk1[' Y']
y_2 = bk2[' Y']
y_3 = bk3[' Y']
y_4 = bk4[' Y']
````

## __Simple Seaborn plot__

````
# Plot using seaborn
_ = plt.figure(figsize=(15, 9)) # (Width x Height)
_ = sns.lineplot(x=lam_1, y=y_1, label='0.1 degree')
_ = sns.lineplot(x=lam_2, y=y_2, label='0.2 degree')
_ = sns.lineplot(x=lam_3, y=y_3, label='2.0 degree')
_ = sns.lineplot(x=lam_4, y=y_4, label='1.0 degree')
_ = plt.xlabel('wavelenth')
_ = plt.ylabel('transmission')
_ = plt.title('transmission vs wavelenth')
_ = plt.legend()
plt.show()
````

## __Plot using ggplot style__

````
# Book1 - 0.1 degree
# Book2 - 0.2 degree
# Book3 - 2.0 degree
# Book4 - 1.0 degree

# plot style
plt.style.use('ggplot')            # Style - 1 ggplot
#sns.set_theme(style="whitegrid")  # style - 2 sns whitegrid(Uncomment it to use the style)

# Plot using seaborn
_ = plt.figure(figsize=(15, 9), facecolor='lightgray') # Remove the facecolor if you need white border
_ = sns.lineplot(x=lam_1, y=y_1, linewidth = 3, color='tab:red', label='0.1 degree')
_ = sns.lineplot(x=lam_2, y=y_2, linewidth = 3, color='tab:blue', label='0.2 degree')
_ = sns.lineplot(x=lam_3, y=y_3, linewidth = 3, color='tab:green', label='2.0 degree')
_ = sns.lineplot(x=lam_4, y=y_4, linewidth = 3, color='tab:orange', label='1.0 degree')
_ = plt.xlabel('wavelenth')
_ = plt.ylabel('transmission')
_ = plt.title('transmission vs wavelenth')
_ = plt.legend()
#plt.savefig('plot1.png') to save the file remove the comment
plt.show()
````

## __sns Whitegrid__

````
# Book1 - 0.1 degree
# Book2 - 0.2 degree
# Book3 - 2.0 degree
# Book4 - 1.0 degree

# plot style
#plt.style.use('ggplot')          # Style - 1 ggplot
sns.set_theme(style="whitegrid")  # style - 2 sns whitegrid

# Plot using seaborn
_ = plt.figure(figsize=(15, 9), facecolor='lightgray') # Remove the facecolor if you need white border
_ = sns.lineplot(x=lam_1, y=y_1, linewidth = 3, palette="tab10", label='0.1 degree') # palette for color
_ = sns.lineplot(x=lam_2, y=y_2, linewidth = 3, palette="tab10", label='0.2 degree')
_ = sns.lineplot(x=lam_3, y=y_3, linewidth = 3, palette="tab10", label='2.0 degree')
_ = sns.lineplot(x=lam_4, y=y_4, linewidth = 3, palette="tab10", label='1.0 degree')
_ = plt.xlabel('wavelenth')
_ = plt.ylabel('transmission')
_ = plt.title('transmission vs wavelenth')
_ = plt.legend()
#plt.savefig('plot1.png') to save the file remove the comment
plt.show()
````
